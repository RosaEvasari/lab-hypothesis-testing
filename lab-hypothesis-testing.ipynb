In this challenge, we will be working with pokemon data. The data can be found here:

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv

```python
#libraries
import pandas as pd
import scipy.stats as st
import numpy as np

```

```python
df_pokemon = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv")
df_pokemon
```

- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

```python
df_pokemon["Type 1"].value_counts()
```

```python
# two-samples t-test

df_dragon = df_pokemon[df_pokemon["Type 1"]=="Dragon"]["HP"]
df_grass = df_pokemon[df_pokemon["Type 1"]=="Grass"]["HP"]
```

```python
# Set the hypothesis --> we want to aim for H1, so, put the H0 as the opposite 

#H0: mu_HP dragron <= mu_HP grass
#H1: mu_HP dragron > mu_HP grass

#significance level = 0.05

st.ttest_ind(df_dragon,df_grass, alternative="greater") 
```

p-value < significant level, this means that we reject the null hypothesis.

- We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.


```python
# Legendary and Non-Legendary pokemon are two independent groups --> Two sample T-test

df_legendary = df_pokemon[df_pokemon["Legendary"]==True]
df_non_legendary = df_pokemon[df_pokemon["Legendary"]==False]

stats = ["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]
```

```python
# Set the hypothesis

#H0: mu_stats_legendary != mu_stats_non_legendary
#H1: mu_stats_legendary = mu_stats_non_legendary
alpha = 0.05

for i in stats:
    t_stat, pvalue = st.ttest_ind(df_legendary[i],df_non_legendary[i], alternative="two-sided")
    print(f'p-value of stat {i}:', {pvalue})
    
    if pvalue < alpha:
        print("Fail to reject null hypothesis.")
    else:
        print("Reject null hypothesis.")
```

```python
# Another approach


df_pokemon['stats'] = df_pokemon[["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]].mean(axis=1)

df_legendary = df_pokemon[df_pokemon["Legendary"]==True]["stats"]
df_non_legendary = df_pokemon[df_pokemon["Legendary"]==False]["stats"] 

```

```python
# Set the hypothesis

#H0: mu_stats_legendary = mu_stats_non_legendary
#H1: mu_stats_legendary != mu_stats_non_legendary

alpha = 0.05

t_stat, pvalue = st.ttest_ind(df_legendary,df_non_legendary, alternative="two-sided")
pvalue
```

p-value < significant level, this means that we reject the null hypothesis.

```python
# Euclidian distance (D) formula: D = ((x2 - x1)^2 + (y2-y1)^2)^0.5

def euclidian_distance_to_school(x1,y1):
    d_school = round((((abs(-118) - abs(x1))**2 + (37-y1)**2)**0.5),2)
    
    return d_school

```

```python

df['d_school'] = df.apply(lambda row: euclidian_distance_to_school(row['longitude'], row['latitude']), axis = 1)
df['d_hospital'] = df.apply(lambda row: euclidian_distance_to_hospital(row['longitude'], row['latitude']), axis = 1)
```

```python
# Euclidian distance (D) formula: D = ((x2 - x1)^2 + (y2-y1)^2)^0.5

def euclidian_distance(x1,y1):
    d_school = round((((abs(-118) - abs(x1))**2 + (37-y1)**2)**0.5),2)
    d_hospital = round((((abs(-122) - abs(x1))**2 + (34-y1)**2)**0.5),2)
    return d_school, d_hospital
```

```python
df[['d_school', 'd_hospital']] = df.apply(lambda row: pd.Series(euclidian_distance(row['longitude'], row['latitude'])), axis=1)
```

```python
df.head(3)
```

```python
df['distance'] = df.apply(lambda row: 'close' if (row['d_school'] < 0.5 or row['d_hospital'] < 0.5) else 'far', axis=1)
df.head(3)
```

```python
# We posit that houses close to either a school or a hospital are more expensive

# Set Hypothesis - Two Sample T-Test

# H0: mu_close_price <= mu_far_price
# H1: mu_close_price > mu_far_price

df_close_price = df[df['distance'] == 'close']['median_house_value']
df_far_price = df[df['distance'] == 'far']['median_house_value']
```

```python
alpha = 0.05

t_stat, pvalue = st.ttest_ind(df_close_price,df_far_price, alternative="greater") 
print('p value is', pvalue)

if pvalue < alpha:
    print("Fail to reject null hypothesis.")
else:
    print("Reject null hypothesis.")
```
